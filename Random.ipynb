```python
from nptdms import TdmsFile
import matplotlib.pyplot as plt
import numpy as np
%%matplotlib inline 
#qt
```

SJSU fastest autoX

```python
tdms_file = TdmsFile.read("C:/Users/willi/Downloads/250620-190319.tdms")
all_groups = tdms_file.groups()
print(all_groups)
data = tdms_file["Data"]
all_group_channels = data.channels()
print(all_group_channels)
Voltage = data['Voltage']
Current = data['Current']
vlist = Voltage[:]
clist = Current[:]
power = []
for char in range(len(vlist)):
    power.append(vlist[char] * clist[char]/1000)
```

```python
plt.plot(power)
```

```python
#each count is 1/100 of a second

#get first run
power = power[10000:15400]
time = np.linspace(0,len(power), len(power))
power = np.array(power)


count = 0
for char in power:
    if char > 56:
        count +=1
print("Time above 56 kw: " , count/100)

slower = np.ma.masked_where(power < 56, power)
supper = np.ma.masked_where(power > 56, power)
plt.plot(time,slower, time, supper)
```
